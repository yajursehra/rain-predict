```python
import pandas as pd
import matplotlib as plt
%matplotlib inline
import numpy as np
```

```python
df=pd.read_csv('weather.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df['RainTomorrow']=df['RainTomorrow'].str.replace('Yes','1')
df['RainTomorrow']=df['RainTomorrow'].str.replace('No','0')
```

```python
df.isnull().sum()
```

```python
df.RainTomorrow=df.RainTomorrow.astype('float32')
```

```python
df['RainToday']=df['RainToday'].str.replace('Yes','1')
df['RainToday']=df['RainToday'].str.replace('No','0')
```

```python
df.RainToday=df.RainToday.astype('float32')
```

```python
df.describe()
```

```python
df.head()
```

```python
fd=df.dropna()
```

```python
fd.head()
```

```python
ff = pd.DataFrame(zip(fd.MinTemp, fd.MaxTemp,fd.Evaporation,fd.Rainfall,fd.Sunshine,fd.WindGustSpeed,fd.WindSpeed9am,fd.WindSpeed3pm,fd.Humidity9am,fd.Humidity3pm,fd.Pressure9am,fd.Pressure3pm,fd.Cloud9am,fd.Cloud3pm,fd.Temp9am,fd.Temp3pm,fd.RainToday,fd.RainTomorrow))
```

```python
ff.columns=['mintemp','maxtemp','evaporation','rainfall','sunshine','windspeed','wind9','wind3','humidity9','humidity3','pressure9','pressure3','cloud9','cloud3','temp9','temp3','raintoday','raintomorrow']
```

```python
ff.head()
```

```python
ff.dtypes
```

```python
ff.head()
```

```python
ff.describe()
```

```python
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
feature_cols = ['mintemp','maxtemp','evaporation','rainfall','sunshine','windspeed','wind9','wind3','humidity9','humidity3','pressure9','pressure3','cloud9','cloud3','temp9','temp3','raintoday']
X = ff[feature_cols] 
y = ff.raintomorrow
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True, feature_names = feature_cols,class_names=['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('diabetes.png')
Image(graph.create_png())
```
